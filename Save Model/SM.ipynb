```python
import numpy as np
from sklearn import linear_model
import pandas as pd
```

```python
df= pd.read_csv('homeprices.csv')
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[5000]])
```

```python
import pickle
```

```python
with open('reg_pickle','wb') as f:
    pickle.dump(reg,f)
```

```python
with open('reg_pickle','rb') as f:
    mp= pickle.load(f)
```

```python
mp.predict([[5000]])
```
