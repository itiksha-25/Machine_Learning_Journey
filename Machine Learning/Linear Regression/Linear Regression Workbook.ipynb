```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df= pd.read_csv('percapita.csv')
df.head(3)
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income')
plt.scatter(df.year, df.pci, color= 'purple', marker= '*')
plt.show()
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['year']],df.pci)
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income')
plt.scatter(df.year, df.pci, color= 'purple', marker= '*')
plt.plot(df.year, reg.predict(df[['year']]), color= 'red')
plt.show()
```
